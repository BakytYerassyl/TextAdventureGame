```python
######################################################################
#          DocString is used as an introduction to the story          #
######################################################################
"""
    DocString:
    
    A) Introduction:
    We all know that professional football(soccer) is a tough place and 
    as long as you are playing in a professional team you have to 
    prove your qualification in every match. This game was based on 
    a fictional story of a player. Imagine you are the prof. player 
    and team wants to replace you next season and today 
    is the last game of the season and it is your last chance. 
    The game consists of 4 stages and does not require any knowledge in football.
   
    
    Stage 1: Receive the ball and make an instant decision
    Stage 2: Proceed to the midfield
    Stage 3: Beat a defender
    Stage 4: Complete your incursion
    
    B) Known Bugs and/or Errors:
    None.
    
"""
from sys import exit
import random


def first_stage():#beginning of the first interactive stage of the game
    print ("""
You received the ball from a teammate, opponents putting a lot of pressure, 2 rival 
players running at you, what will you do?
1. Just kick the ball hard and hope for luck
2. Give the ball to the mate who is under the pressure
3. Try to squeeze through rivals \n""") #\n been used every time to make an output readable for the user
#given several options to choose    
    break_pressure = input(prompt = "> \n")
#input of any option user decided
    if "1" in break_pressure or "A" in break_pressure or "kick" in break_pressure.lower() or "hope" in break_pressure.lower() or "luck" in break_pressure.lower():
#method lower() been used in every possible user option to provide for code no sensibility for any inputs(small letter or capital)
        x = random.randint(1,10) # this is a self-made randomizer function to gives a bit more than 1% chance of success
        y = random.randint(1,10)
        if x == y: #so here we have 1 out of 100, if the numbers are equal output = success
            print ("What a magic you scored from your own side of a pitch! \n")
            victory()#if the user is lucky enough success in this stage takes the user directly to the end of the story
        else:#because it is obvious that it is hard to score from so far, only 1% of success
            print ("You lost the ball and rivals controlled the ball until the match ended as 0 - 0 \n")
            draw()#there are few possible results, draw one of them, even failure brings the user to the final part of the game
    elif "2" in break_pressure or "B" in break_pressure or "give" in break_pressure.lower() or "pass" in break_pressure.lower() or "mate" in break_pressure.lower():
        print ("""That was a bad decision. Your mate lost the ball under the pressure,
so rivals scored easily. Final score is 0 - 1 \n""")
        bad_end()#one of possible ends of the story
    elif "3" in break_pressure or "C" in break_pressure or "try" in break_pressure.lower() or "squeeze" in break_pressure.lower():
        print ("Nice one, no one expected that from you! \n")
        teamplay()
#if the user will do right decision next stage begins
    else: #else function was added to catch all invalid inputs and don't lead the code to crash on simple things, 
#utility of it is that you don't need to start the game again because of a small typo or etc.
        print("Invalid entry, please try again")
        first_stage()
#else clause been added through all code to catch all input errors and bring the user to the beginning of the stage        

def teamplay():
    print (f""" 
Okay {status}{user_name.capitalize()}, you took an initiative, but now rivals 
are expecting from you that you can go dribbling, what is your next step?
1. Kick the ball hard from the center of the pitch and hope for luck
2. Pass the ball to the closest mate and run forward
3. Try to escape from the pressure of next two rivals \n""") #f been used to input a function inside print
    teamplay = input(prompt = "> \n")
    if "1" in teamplay or "kick" in teamplay.lower() or "hope" in teamplay.lower() or "luck" in teamplay.lower() or "center" in teamplay.lower():
        a = random.randint(1,8) #on more randomizer with the opportunity of success higher than in previuos one (because to score from the center is a still hard thing) 
        b = random.randint(1,8)
        if a == b:
            print ("What a beauty! You scored from the center of a pitch! \n")
            victory()
        else:
            print ("You lost the ball and rivals controlled the ball until match ended as 0 - 0 \n")
            draw()
    elif "2" in teamplay or "pass" in teamplay.lower() or "run forward" in teamplay.lower() or "mate" in teamplay.lower():
        print("""
Good decision, football is a team game, you passed the ball and ran forward, your
mate gave the ball back and you suddenly ended up close to rival's goal, only one 
defender left to beat \n""")
        last_def()#takes the user to the next stage of the game
    elif "3" in teamplay or "try" in teamplay.lower() or "escape" in teamplay.lower():
        print("""It was the wrong decision, rivals knew that you can try to pass them again
and with joint efforts took the ball away. Finally match ended 0 - 0 \n""")
        draw()
    else:
        print("Invalid input, please try again")
        teamplay()

def last_def():
    defenders_move_list = ["left", "right", "center"] #list of possible moves been added to bring more varieties for the game
    defenders_moves = 1 #defender has only 1 opportunity to catch a player and mostly defenders are guessing where will forward move, 
#beacuse player who is running has advantage in front of defender who is in a static position
    while defenders_moves > 0:#the defender has only 1 chance to catch the user
        defenders_move = random.choice(defenders_move_list) #phyton randomly choosing next defenders move from a list 
        input(prompt = "<Press enter to see what will defender do>\n")
        
        
        if defenders_move == defenders_move_list[0]: #note lists in phyton starts with 0
            print("""
The defender is guessing your next step and started to move left. What will you do to pass him?
1. Quickly run left
2. Make a fake shoot and run right \n""")
            move_1 = input (prompt = "> \n")
            if "1" in move_1 or "left" in move_1.lower():
                print ("Defender guessed right and took the ball away, you lost the ball \n")
                draw()
            elif "2" in move_1 or "right" in move_1.lower() or "fake shot" in move_1.lower():
                print ("You passed the last defender successfully and getting ready to shoot \n")
                defenders_moves -=1
#as been said before defender has only 1 chance
                finish()
            else:
                print("Invalid input, try again")
                last_def()
        elif defenders_move == defenders_move_list[1]:
            print("""Defender is guessing your next step and started to move right.
What you will do to pass him?
1. Quickly run right
2. Make a fake shoot and run left \n""")
            move_2 = input (prompt = "> \n")
            if "1" in move_2 or "right" in move_2.lower():
                print ("Defender guessed right and took the ball away, you lost the ball \n")
                draw()
            elif "2" in move_2 or "left" in move_2.lower() or "fake shot" in move_2.lower():
                print ("You passed the last defender successfully and getting ready to shoot \n")
                defenders_moves -=1
                finish()#takes the user to the next stage
            else:
                print("Invalid input, try again")
                last_def()
        elif defenders_move == defenders_move_list[2]: #the situation when defender late with timing and can't do anything
                print ("""Defender is staying at center.
You can run in any direction and he will not catch you anyway
1. Quickly run right
2. Quickly run left \n""")
                move_3 = input (prompt = "> \n")
                if "1" in move_3 or "2" in move_3 or "right" in move_3.lower() or "left" in move_3.lower() or "run" in move_3.lower():
                    print("You passed the last defender successfully and getting ready to shoot \n")
                    defenders_moves -=1
                    finish()#takes the user to the next stage
                else:
                    print ("Invalid input, please try again")
                    last_def()

def finish():
    input(prompt = "<Press enter to continue>\n")
    print("""
The last action to take! Which side of the goal are you going to shoot? 
Goalkeeper is still here and he is trying to save his net
1. Left corner
2. Center
3. Right corner
4. Pass to your mate, who is opened and will score to the empty target \n""")
    shot = input (prompt = "> \n") #randomizer again if player shoots left he/she has only 25% chance to score
    if "1" in shot or "left" in shot.lower():
        c = random.randint(1,2)
        d = random.randint(1,2)
        if c == d:
            print("""
   _____                                     _ _           
  / ____|                                   | | |          
 | |  __  ___   ___   ___   ___   ___   ___ | | |          
 | | |_ |/ _ \ / _ \ / _ \ / _ \ / _ \ / _ \| | |          
 | |__| | (_) | (_) | (_) | (_) | (_) | (_) | |_|          
  \_____|\___/ \___/ \___/ \___/ \___/_\___/|_(_)_   ____  
  / ____|/ __ \| |        /\    |___  / __ \ / __ \ / __ \ 
 | |  __| |  | | |       /  \      / / |  | | |  | | |  | |
 | | |_ | |  | | |      / /\ \    / /| |  | | |  | | |  | |
 | |__| | |__| | |____ / ____ \  / /_| |__| | |__| | |__| |
  \_____|\____/|______/_/    \_\/_____\____/ \____/ \____/ 
  """)
            victory()
        else:
            print ("Oh noooo! Keeper guessed right and cought the ball \n")
            draw()
    elif "2" in shot or "center" in shot.lower(): #bad decision to shoot in the middle where is keeper, only 11% chance of success
        e = random.randint(1,3)
        f = random.randint(1,3)
        if e == f:
            print("""
   _____                                     _ _           
  / ____|                                   | | |          
 | |  __  ___   ___   ___   ___   ___   ___ | | |          
 | | |_ |/ _ \ / _ \ / _ \ / _ \ / _ \ / _ \| | |          
 | |__| | (_) | (_) | (_) | (_) | (_) | (_) | |_|          
  \_____|\___/ \___/ \___/ \___/ \___/_\___/|_(_)_   ____  
  / ____|/ __ \| |        /\    |___  / __ \ / __ \ / __ \ 
 | |  __| |  | | |       /  \      / / |  | | |  | | |  | |
 | | |_ | |  | | |      / /\ \    / /| |  | | |  | | |  | |
 | |__| | |__| | |____ / ____ \  / /_| |__| | |__| | |__| |
  \_____|\____/|______/_/    \_\/_____\____/ \____/ \____/ 
  """)
            victory()
        else:
            print ("Oh noooo! Keeper guessed right and cought the ball \n")
            draw() 
    elif "3" in shot or "right" in shot.lower():
        print("""
   _____                                     _ _           
  / ____|                                   | | |          
 | |  __  ___   ___   ___   ___   ___   ___ | | |          
 | | |_ |/ _ \ / _ \ / _ \ / _ \ / _ \ / _ \| | |          
 | |__| | (_) | (_) | (_) | (_) | (_) | (_) | |_|          
  \_____|\___/ \___/ \___/ \___/ \___/_\___/|_(_)_   ____  
  / ____|/ __ \| |        /\    |___  / __ \ / __ \ / __ \ 
 | |  __| |  | | |       /  \      / / |  | | |  | | |  | |
 | | |_ | |  | | |      / /\ \    / /| |  | | |  | | |  | |
 | |__| | |__| | |____ / ____ \  / /_| |__| | |__| | |__| |
  \_____|\____/|______/_/    \_\/_____\____/ \____/ \____/ 
  """)
        victory()
    elif "4" in shot or "pass" in shot.lower() or "mate" in shot.lower():
        print ("Good decision! Your teammate scored! \n")
        normal_end()
    else:
        print("Invalid entry, keep calm and try again \n")
        finish()
#the final part of the game user has 4 possible ends of the game - draw, loose, normal and win        
def victory():
    print("Great job! You have done it!")
    input(prompt = "<Press enter to know the end of the story> \n")
    print(f"""
Based on the last game results {status}{user_name.capitalize()} showed excellent
skills and got promoted. The club decided to keep him for the next season and 
managers decided that {position.lower()} position does not need any improvements.
After that historical goal {status}{user_name.capitalize()} became an idol for the fans!
    
     _______. __    __    ______   ______  _______     _______.     _______.
    /       ||  |  |  |  /      | /      ||   ____|   /       |    /       |
   |   (----`|  |  |  | |  ,----'|  ,----'|  |__     |   (----`   |   (----`
    \   \    |  |  |  | |  |     |  |     |   __|     \   \        \   \    
.----)   |   |  `--'  | |  `----.|  `----.|  |____.----)   |   .----)   |   
|_______/     \______/   \______| \______||_______|_______/    |_______/    
                                                                          """)
    exit(0)
    
def normal_end():
    print("Great job! You have done it!")
    input(prompt = "<Press enter to know the end of the story> \n")
    print(f"""
Based on the last game results {status}{user_name.capitalize()} could not show his
high class said the coach after the game, he hasn't been sacked but he lost his 
line up place and got pay cut. His team signed another player for his {position.lower()} position, 
even though, he tried to show his maximum. Only some of the advanced fans found
something special in his last game performance.

The end!""")
    exit(0)
      
    
def draw():
    print("Match ended, the score is 0 - 0, you failed your last chance to show yourself")
    input(prompt = "<Press enter to know the end of the story> \n")
    print(f"""
Based on the last game results {status}{user_name.capitalize()} showed less than
expected, he hasn't been sacked but he lost his lineup place and got pay cut. 
His team signed another player for his {position.lower()} position, even though,
he tried to show his maximum.
    
   _____                                            _ 
  / ____|                                          | |
 | |  __  __ _ _ __ ___   ___    _____   _____ _ __| |
 | | |_ |/ _` | '_ ` _ \ / _ \  / _ \ \ / / _ \ '__| |
 | |__| | (_| | | | | | |  __/ | (_) \ V /  __/ |  |_|
  \_____|\__,_|_| |_| |_|\___|  \___/ \_/ \___|_|  (_)
                                                      
                                                      """)
    exit(0)
    
def bad_end():
    print("Match ended, you couldn't help your team, moreover you set up your mate")
    input(prompt = "<Press enter to know the end of the story> \n")
    print(f"""
{status}{user_name.capitalize()} was the worst player on the pitch and made a crucial
mistake which led to a conceded goal and lost match. Right after the game 
{status}{user_name.capitalize()} was fired. The next day the club signed another player for
his {position.lower()} position and won a trophy next season.
    
   _____                                            _ 
  / ____|                                          | |
 | |  __  __ _ _ __ ___   ___    _____   _____ _ __| |
 | | |_ |/ _` | '_ ` _ \ / _ \  / _ \ \ / / _ \ '__| |
 | |__| | (_| | | | | | |  __/ | (_) \ V /  __/ |  |_|
  \_____|\__,_|_| |_| |_|\___|  \___/ \_/ \___|_|  (_)
  
  """)
    exit(0)

#game starter code in bottom because phyton reads a code from top to bottom, first we need to identify all functions
# and then use them, starter haven"t beed inserted as " def start()" function because we use {status} and {user_name}
# inputs through all code, and "def start()" function will not allow us to call it in other functions such as in "def bad_end()"
input (prompt = "<Press enter to start>\n")#prompt explicit argument been added through all code inputs 
user_name = input (prompt = """What is your name?
Please be honest and write your name correctly with capital letter\n""")
position = input (prompt = """What is your position? e.g.
- Goalkeeper
- Defender
- Midfielder
- Forward\n""")
#position option does not have restrictions, because not everyone familiar with soccer so user can choose any position even for example "bench"

status = input (prompt = "If you are professor please press Y, if not just skip <press enter>")
if status == 'Y' or status == "y" or "es" in status: #variable "status" been added to make it more for better using experiance for prof. Chase
    status = "Professor "
else:
    status = "" #if status provided as anything else than "yes" code will not count it
print (f"Welcome to professional football {status}{user_name.capitalize()}!")

input (prompt = "<Press enter to dive in to the situation>")
print (f"""

Quick description to bring you up to date, as been said before it is your last 
chance to show yourself, but the game is almost finished, the score is 0-0
and its 87th minute, you are already thinking what you will do after 
they fire you, but suddenly your partner passed you the ball and you 
recalled the video you have seen on youtube yesterday how one player had beaten all 
rivals team and scored, so you decided to do that, common {status}{user_name.capitalize()}!!!""")
#to print user name with capital letter capitalize() method been used through all code
input (prompt = "<Press enter to continue>")
first_stage()#takes the user to the first stage of the game
```
